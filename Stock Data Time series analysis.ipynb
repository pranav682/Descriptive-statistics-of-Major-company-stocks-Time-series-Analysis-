# Use Rolling Method to compute moving Averages

```python
Amzn['MA_9']=Amzn.Amzn_close.rolling(9).mean().shift()
Amzn['MA_21']=Amzn.Amzn_close.rolling(21).mean().shift()
Amzn['MA_9'].head(20)
```

# Plot Data and Moving Averages

```python
plt.figure(figsize =(10,5))
plt.grid(True)
plt.plot(Amzn['Date'], Amzn.Amzn_close,label="Amazon") 
plt.plot(Amzn['Date'], Amzn['MA_9'],label="MA 9 day")
plt.plot(Amzn['Date'], Amzn['MA_21'],label="MA 21 day")
plt.legend(loc=2)
```

```python
googl['MA_9']=googl.googl_close.rolling(9).mean().shift()
googl['MA_21']=googl.googl_close.rolling(21).mean().shift()
plt.figure(figsize =(10,5))
plt.grid(True)
plt.plot(googl['Date'], googl.googl_close,label="Google") 
plt.plot(googl['Date'], googl['MA_9'],label="MA 9 day")
plt.plot(googl['Date'], googl['MA_21'],label="MA 21 day")
plt.legend(loc=2) 
```

```python
mkst['MA_9']=mkst.mkst_close.rolling(9).mean().shift()
mkst['MA_21']=mkst.mkst_close.rolling(21).mean().shift()
plt.figure(figsize =(10,5))
plt.grid(True)
plt.plot(mkst['Date'], mkst.mkst_close,label="Microsoft") 
plt.plot(mkst['Date'], mkst['MA_9'],label="MA 9 day")
plt.plot(mkst['Date'], mkst['MA_21'],label="MA 21 day")
plt.legend(loc=2)
```

*** As we can see above in the plot, Due to recession during 2008 all the stockprices dropped rapidly for next few years. but Amazon being a small company by that time didnt have to face large losses as compared to what other established aompanies have faced ***

# Computing log change(Instantaneous rate of return)

```python
Amzn['change'] = np.log(Amzn['Amzn_close']/Amzn['Amzn_close'].shift()) 
googl['change'] = np.log(googl['googl_close']/googl['googl_close'].shift()) 
mkst['change'] = np.log(mkst['mkst_close']/mkst['mkst_close'].shift())
```

```python
plt.plot(Amzn['Date'],Amzn['change'])
```

```python
plt.plot(googl['Date'],googl['change'])
```

```python
plt.plot(mkst['Date'],mkst['change'])
```

*** plot above reveals noisy data centered around zero ***

# Compute rolling historical volatility offset using .shift() method

```python
Amzn['volatility'] = Amzn.change.rolling(21).std().shift()
plt.plot(Amzn['Date'],Amzn['volatility'])
```

```python
googl['volatility'] = googl.change.rolling(21).std().shift()
plt.plot(googl['Date'],googl['volatility'])
```

```python
mkst['volatility'] = mkst.change.rolling(21).std().shift()
plt.plot(mkst['Date'],mkst['volatility'])
```

# Some price Analysis

```python
#offsetting expected change 1-dy to prevent "forecasting" current day.
Amzn['Actual_change'] = Amzn.Amzn_close - Amzn.Amzn_close.shift(1) 
googl['Actual_change'] = googl.googl_close - googl.googl_close.shift(1)
mkst['Actual_change'] = mkst.mkst_close - mkst.mkst_close.shift(1)

Amzn.head()
```

```python
Amzn['exp_change'] =  Amzn.Amzn_close.shift(1)*Amzn['volatility'] 
googl['exp_change'] =  googl.googl_close.shift(1)*googl['volatility']
mkst['exp_change'] =  mkst.mkst_close.shift(1)*mkst['volatility']
```

```python
#running more than once will give progressively smaller datasets
Amzn = Amzn.iloc[1:]
googl = googl.iloc[1:]
mkst = mkst.iloc[1:]
```

```python
#Magnitude in standard deviation 
Amzn['Magnitude'] = Amzn['Actual_change']/Amzn['exp_change']
googl['Magnitude'] = googl['Actual_change']/googl['exp_change']
mkst['Magnitude'] = mkst['Actual_change']/mkst['exp_change']
Amzn.describe()
```

```python
df = googl['Magnitude'].dropna().to_frame()
df1 = Amzn['Magnitude'].dropna().to_frame()
df2 = mkst['Magnitude'].dropna().to_frame()
```
